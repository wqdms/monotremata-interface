```python
from interactive import interface
from monotremata import models
MH = interface.ModelHandler()
```

```python
MH.get_model("Organization").objects.all()    
```

```python
unep = MH.update_create("Organization",{"name":"UNEP","tag":"v0"})
wqdms = MH.update_create("Project",{"name":"wqdms","tag":"v0"})
sdg632 = MH.update_create("Application",{"name":"sdg632","tag":"v0"})
nationalwq = MH.update_create("PresetModel",{"name":"NationalWaterQuality","tag":"v0"})
document = MH.get_model("Document").objects.first()
domain = MH.update_create("Domain",{"name":"test","tag":"v0"})
namespace = MH.update_create("Namespace",{"name":"demo","tag":"v0"})
deployment = MH.update_create("Deployment",{
    "name":"demo",
    "tag":"v0",
    "domain":domain,
    "namespace":namespace,
    "organization": unep,
    "project":wqdms,
    "lifecycle_hook":"deploy"
})
deployment.applications.add(sdg632)
unep.projects.add(wqdms)
wqdms.applications.add(sdg632)
sdg632.preset_models.add(nationalwq)
```
